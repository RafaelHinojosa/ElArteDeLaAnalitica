```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import figure
```

```python
# 1. Cargar los datos
songs = pd.read_csv("../data/songs.csv")
```

```python
# Select Data (key and tempo)
test_key_tempo = songs[["key", "tempo"]]
test_key_tempo = test_key_tempo.dropna(axis=0, how='any')
```

```python
# Kmeans algorithm
kmeans = KMeans(n_clusters=7).fit(test_key_tempo)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predict class from test
prediction = kmeans.predict(test_key_tempo)
```

```python
# Predict nuevo dato
data = {'key': ['0'], 'tempo': [120]}
newDF = pd.DataFrame(data)
print(kmeans.predict(newDF))
```

```python
# Graph scatter plot
plt.scatter(songs["key"], songs["tempo"], c=prediction)

for i in range(len(centroids)):
    plt.scatter(centroids[i][0], centroids[i][1], marker="*", c="red")

plt.xlabel('Key')
plt.ylabel('Tempo')
plt.title('Clasificación de canciones por key y tempo')
plt.show()
```
